```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,color="#1E8591",alpha=0.3)
benimEksen.plot(numpyDizisi2,numpyDizisi1,color="#C38442")
```

```python
(yenşFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1 +2,color="blue",linewidth=5.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1 +3,color="purple",linewidth=1.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1 +4,color="#C38442",linestyle=":")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 +5,color="#C38442",linestyle="-.")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 +6,color="#C38442",linestyle="--")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 +7,color="#000000",linestyle="--",marker="o",markerfacecolor="red")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 +8,color="#000000",linestyle="--",marker="+",markersize=8)
```
