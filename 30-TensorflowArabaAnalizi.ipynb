```python
dataFrame = dataFrame.drop("transmission",axis=1) # y ulaşmak istediğimiz şey
```

```python
y = dataFrame["price"].values #diziye çevirdik
x= dataFrame.drop("price",axis=1).values #price dışındakileri aldık
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10) # random state rastgele bölüyor train ve testleri
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #modelimizi oluşturuyoruz
from tensorflow.keras.layers import Dense #layerimizi oluşturuyoruz
```

```python
x_train.shape #içerisinde 9090 veri var ve 5 tane de özellik var. katmanları 5 ten başlatsak iyi olur.
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1)) #çıkış katmanı 1 tane 

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
